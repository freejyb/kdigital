## ì—°ë³µë¦¬ 10% 2ë…„ë§Œê¸° , 10% ì´ììœ¨

```python
def calculate_compound_interest(principal, rate, time, n=1):
    """
    ë³µë¦¬ ê³„ì‚° í•¨ìˆ˜

    :param principal: ì›ê¸ˆ (Principal)
    :param rate: ì—°ì´ììœ¨ (Annual interest rate)
    :param time: ê¸°ê°„ (Time the money is invested for)
    :param n: ì´ìê°€ ë³µë¦¬ë¡œ ì ìš©ë˜ëŠ” íšŸìˆ˜ (Number of times interest applied per time period)
    :return: ë§Œê¸° ì‹œ ë°›ëŠ” ê¸ˆì•¡ (Amount)
    """
    amount = principal * (1 + rate / n) ** (n * time)
    return amount

# ì˜ˆì œ íŒŒë¼ë¯¸í„°
principal = 10000000  # ì›ê¸ˆ (10,000,000ì›)
rate = 0.10  # ì—°ì´ììœ¨ (10%)
time = 2  # ê¸°ê°„ (2ë…„)

# ë§Œê¸° ê¸ˆì•¡ ê³„ì‚°
final_amount = calculate_compound_interest(principal, rate, time)
print(f"ë§Œê¸° ì‹œ ë°›ëŠ” ê¸ˆì•¡: {final_amount:.2f}ì›")
```

### ë§¤ì›” 1000ë‹¬ëŸ¬, 10ë…„ê°„ ì—°ê¸ˆ ìˆ˜ë ¹, í• ì¸ìœ¨ 5%, ì§€ê¸‰ê¸°ê°„ 10ë…„

```python
def calculate_annuity_pv(payment, rate, periods):
    """
    ì—°ê¸ˆì˜ í˜„ì¬ ê°€ì¹˜ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜.

    :param payment: ê° ê¸°ê°„ë§ˆë‹¤ ë°›ëŠ” ê¸ˆì•¡ (Payment per period)
    :param rate: í• ì¸ìœ¨ (Discount rate)
    :param periods: ì´ ì§€ê¸‰ ê¸°ê°„ ìˆ˜ (Number of periods)
    :return: í˜„ì¬ ê°€ì¹˜ (Present Value)
    """
    pv = payment * (1 - (1 + rate) ** -periods) / rate
    return pv

# ì˜ˆì œ íŒŒë¼ë¯¸í„°
payment = 1000  # ê° ê¸°ê°„ë§ˆë‹¤ ë°›ëŠ” ê¸ˆì•¡ (ì˜ˆ: ì›” 1000ë‹¬ëŸ¬)
rate = 0.05 / 12  # ì›” í• ì¸ìœ¨ (ì—° 5%ë¥¼ ì›”ë¡œ í™˜ì‚°)
periods = 10 * 12  # ì´ ì§€ê¸‰ ê¸°ê°„ ìˆ˜ (10ë…„ì„ ì›”ë¡œ í™˜ì‚°)

# í˜„ì¬ ê°€ì¹˜ ê³„ì‚°
present_value = calculate_annuity_pv(payment, rate, periods)
print(f"ì—°ê¸ˆì˜ í˜„ì¬ ê°€ì¹˜: {present_value:.2f}ë‹¬ëŸ¬")
```

```python
def calculate_annuity_fv(payment, rate, periods):
    """
    ì—°ê¸ˆì˜ ë¯¸ë˜ ê°€ì¹˜ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜.

    :param payment: ê° ê¸°ê°„ë§ˆë‹¤ ë°›ëŠ” ê¸ˆì•¡ (Payment per period)
    :param rate: í• ì¸ìœ¨ (Discount rate)
    :param periods: ì´ ì§€ê¸‰ ê¸°ê°„ ìˆ˜ (Number of periods)
    :return: ë¯¸ë˜ ê°€ì¹˜ (Future Value)
    """
    fv = payment * ((1 + rate) ** periods - 1) / rate
    return fv

# ì˜ˆì œ íŒŒë¼ë¯¸í„°
payment = 1000  # ê° ê¸°ê°„ë§ˆë‹¤ ë°›ëŠ” ê¸ˆì•¡ (ì˜ˆ: ì›” 1000ë‹¬ëŸ¬)
rate = 0.05 / 12  # ì›” í• ì¸ìœ¨ (ì—° 5%ë¥¼ ì›”ë¡œ í™˜ì‚°)
periods = 10 * 12  # ì´ ì§€ê¸‰ ê¸°ê°„ ìˆ˜ (10ë…„ì„ ì›”ë¡œ í™˜ì‚°)

# ë¯¸ë˜ ê°€ì¹˜ ê³„ì‚°
future_value = calculate_annuity_fv(payment, rate, periods)
print(f"ì—°ê¸ˆì˜ ë¯¸ë˜ ê°€ì¹˜: {future_value:.2f}ë‹¬ëŸ¬")
```
