# NumPy


## 개괄 이해

```python
# NumPy : Numerical Python
# 다차원(n차원 배열 생성 : 1차원, 2차원..) 배열(array) 연산기능을 모아둔 패키지
# 넘파이(NumPy)**는 파이썬에서 수치 계산을 위한 가장 기본적인 라이브러리
# 다차원 배열 객체와 이를 효율적으로 다루기 위한 다양한 함수들을 제공합니다. 
# 넘파이는 빠르고 유연한 배열 연산을 통해 수학, 과학, 공학 계산 등에 많이 사용되며, 다른 데이터 분석 및 머신러닝 라이브러리(예: Pandas, SciPy, Scikit-learn 등)의 기반이 되기도 합니다.

# 벡터, 행렬, 다차원 배열을 위한 파이썬 객체 및 이 객체들을 이용한 수치연산, 통계 계산 등의 기능
# 행렬, 배열의 수학연산을 벡터연산(일종의 병렬 연산)을 하여 성능을 개선 (연산속도 측면)
```

```python
# http://www.numpy.org/
```

## 넘파이의 주요 특징과 기능

```python
# 1. N차원 배열 객체 (ndarray):
# 넘파이의 핵심 데이터 구조로, 동일한 데이터 타입의 값들이 담긴 다차원 배열입니다. 
# 리스트와 유사하지만, 배열 크기가 고정되어 있고, 수치 계산에 훨씬 효율적입니다.
# 2. 빠른 연산 속도:
# 넘파이는 C 언어로 구현되어 있어 파이썬의 기본 리스트보다 훨씬 빠르게 수학적 연산을 수행할 수 있습니다.
# 3. 벡터화 연산:
# 반복문 없이 배열 전체에 대한 연산을 한 번에 수행할 수 있습니다. 
# 이를 통해 코드가 간결해지고 연산 속도가 빨라집니다.
# 4. 브로드캐스팅(Broadcasting):
# 크기가 다른 배열 간의 연산을 자동으로 맞춰 수행할 수 있습니다. 
# 예를 들어, 배열과 스칼라 값 간의 연산에서 배열의 각 요소에 스칼라 값이 적용됩니다.
# 5. 다양한 수학 함수:
# 선형대수, 통계, 푸리에 변환 등의 다양한 수학 함수들을 제공합니다.
# 6. 난수 생성:
# 다양한 분포에서 난수를 생성할 수 있어, 시뮬레이션이나 머신러닝에서 많이 활용됩니다.
# 7.파일 입출력:
# 배열 데이터를 바이너리 형식이나 텍스트 형식(CSV, TSV 등)으로 저장하거나 불러올 수 있습니다.
```

## 배열이란?

```python
# 데이터 처리를 효율적으로 하기 위한 자료구조
# 배열 안에는 숫자나 문자열  모두를 담을 수 있다. 단, 배열 안에서는 동일 종류의 데이터만 담을 수 있다.
# - 리스트는 다른 유형의 데이터를 담을 수 있지만 배열은 불가
# 리스트 : [1,2,3] , [안산, 대학교, 금융], [1,2, 안산, 4]
# 배열 : ([1,2,3]) , ([안산, 대학교, 금융])
```

```python
# 배열의 종류(차원: Dimension ) : 1차원, 2차원, 3차원
# n차원의 배열 (ndarray)
# 1) 1차원 배열 : 삼성전자  주가 ([61000,59000,56500]) # 리스트와 유사하나 , 연산 방법의 차이로 속도가 빠르다.
# 2) 2차원 배열 : 삼성전자 주가, 네이버 주가 ( [61000,59000,56500], [ 200000, 210000, 2250000])
# 3) 3차원 배열 : 2차원 배열이 수개 모여 있는 형태 
```

## 배열 만들기 

```python
# 배열을 만들기 위한 함수
# 1) np.array  : 초기 값을 지정하여 배열 만들기
# 2) np.arrange : 순차적인 숫자 이용 배열 만들기
# 3) np. rand : 랜덤하게 숫자 배열 만들기
```

### np.array

```python
# 만드는 방법
# np.array ([나열된 데이터 객체])
```

```python
# 넘파이 불러오기
import numpy as np
# np.array ([나열된 데이터 객체])
# stock1 리스트 데이터를 배열 형태로 변환

stock1=[15000, 14500, 16000, 17000]
arr1=np.array([stock1])
arr1
```

### np.arange

## 배열 탐색하기

```python
# 배열의 타입을 확인하는 경우
# shape : 배열 형태 확인하는 경우
# dtype : 배열 데이터 타입 확인하는 경우
# astype : 배열의 데이터 타입 변경하기
# 인덱스 범위 지정 : 배열 안 데이터 일부 만 보는 경우
```

### shape : 배열 형태 확인하는 경우

```python
arr=np.arange(10)
arr
```

```python
# arr 에 대하여 배열 형태 확인하는 경우
arr.shape
# 결과 10개 값인 1차원 배열
```

### dtype : 배열 데이터 타입 확인하는 경우 

```python
# arr 에 대하여 배열에 포함된 데이터 타입 확인하는 경우
arr.dtype
# 결과 정수 값으로 이루어 짐.
```

### astype : 배열의 데이터 타입 변경하기

```python
# 문자형으로 데이터 타입을 변경하는 경우
arr1=arr.astype(str)
# 문자형으로 변경됨 : '' 안에 숫자가 있다.
arr1
```
