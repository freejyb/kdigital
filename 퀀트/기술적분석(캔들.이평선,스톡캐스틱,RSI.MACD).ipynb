### SMA, WMA, EMA 

```python
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240601"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 단순 이동 평균선 계산
df['SMA_20'] = df['종가'].rolling(window=20).mean()
df['SMA_5'] = df['종가'].rolling(window=5).mean()

# 가중 이동 평균선 계산
def weighted_moving_average(data, window):
    weights = pd.Series(range(1, window + 1))
    return data.rolling(window).apply(lambda prices: (prices * weights).sum() / weights.sum(), raw=True)

df['WMA_20'] = weighted_moving_average(df['종가'], 20)
df['WMA_5'] = weighted_moving_average(df['종가'], 5)

# 지수 이동 평균선(EMA) 계산
df['EMA_20'] = df['종가'].ewm(span=20, adjust=False).mean()
df['EMA_5'] = df['종가'].ewm(span=5, adjust=False).mean()

# 주가 및 이동 평균선 그래프 그리기
plt.figure(figsize=(14, 7))

# 단순 이동 평균선
plt.plot(df.index, df['종가'], label='Samsung Electronics Close Price', color='blue')
plt.plot(df.index, df['SMA_20'], label='20-Day SMA', color='orange')
plt.plot(df.index, df['SMA_5'], label='5-Day SMA', color='green')

# 가중 이동 평균선
plt.plot(df.index, df['WMA_20'], label='20-Day WMA', color='red')
plt.plot(df.index, df['WMA_5'], label='5-Day WMA', color='purple')

# 지수 이동 평균선
plt.plot(df.index, df['EMA_20'], label='20-Day EMA', color='brown')
plt.plot(df.index, df['EMA_5'], label='5-Day EMA', color='pink')

plt.title('Samsung Electronics (005930.KS) Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend()
plt.grid(True)
plt.show()
```

## 볼린저밴드

### 중심선 : 20MA

```python
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240601"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터프레임의 인덱스를 날짜 형식으로 변환
df.index = pd.to_datetime(df.index)

# 이동 평균과 표준 편차 계산
window = 20
df['SMA'] = df['종가'].rolling(window=window).mean()
df['STD'] = df['종가'].rolling(window=window).std()

# 볼린저 밴드 계산
df['Upper Band'] = df['SMA'] + (df['STD'] * 2)
df['Lower Band'] = df['SMA'] - (df['STD'] * 2)

# 볼린저 밴드 시각화
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['종가'], label='Close Price', color='blue')
plt.plot(df.index, df['SMA'], label='20-Day SMA', color='orange')
plt.plot(df.index, df['Upper Band'], label='Upper Band', color='green')
plt.plot(df.index, df['Lower Band'], label='Lower Band', color='red')
plt.fill_between(df.index, df['Upper Band'], df['Lower Band'], color='grey', alpha=0.1)
plt.title('Samsung Electronics (005930.KS) Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

## Stochastic

### 5일

```python
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240601"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터프레임의 인덱스를 날짜 형식으로 변환
df.index = pd.to_datetime(df.index)

# 스토캐스틱 오실레이터 계산 함수
def calculate_stochastic_oscillator(df, period=5):
    df['L5'] = df['저가'].rolling(window=period).min()
    df['H5'] = df['고가'].rolling(window=period).max()
    df['%K'] = (df['종가'] - df['L5']) * 100 / (df['H5'] - df['L5'])
    df['%D'] = df['%K'].rolling(window=3).mean()
    return df

# 스토캐스틱 오실레이터 계산 (5일 기준)
df = calculate_stochastic_oscillator(df, period=5)

# 스토캐스틱 오실레이터 시각화
plt.figure(figsize=(12, 6))

# 주가 그래프
plt.subplot(2, 1, 1)
plt.plot(df.index, df['종가'], label='Close Price', color='blue')
plt.title('Samsung Electronics (005930.KS) Close Price and 5-Day Stochastic Oscillator')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)

# 스토캐스틱 오실레이터 그래프
plt.subplot(2, 1, 2)
plt.plot(df.index, df['%K'], label='%K', color='red')
plt.plot(df.index, df['%D'], label='%D', color='green')
plt.axhline(y=80, color='grey', linestyle='--')
plt.axhline(y=20, color='grey', linestyle='--')
plt.title('5-Day Stochastic Oscillator')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## MACD (Moving Average Convergence Divergence)

```python
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240601"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터프레임의 인덱스를 날짜 형식으로 변환
df.index = pd.to_datetime(df.index)

# MACD 계산 함수
def calculate_macd(df, short_window=12, long_window=26, signal_window=9):
    df['EMA12'] = df['종가'].ewm(span=short_window, adjust=False).mean()
    df['EMA26'] = df['종가'].ewm(span=long_window, adjust=False).mean()
    df['MACD'] = df['EMA12'] - df['EMA26']
    df['Signal Line'] = df['MACD'].ewm(span=signal_window, adjust=False).mean()
    return df

# MACD 계산
df = calculate_macd(df)

# MACD 시각화
plt.figure(figsize=(12, 8))

# 주가 그래프
plt.subplot(3, 1, 1)
plt.plot(df.index, df['종가'], label='Close Price', color='blue')
plt.title('Samsung Electronics (005930.KS) Close Price and MACD')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)

# MACD 및 Signal Line 그래프
plt.subplot(3, 1, 2)
plt.plot(df.index, df['MACD'], label='MACD', color='red')
plt.plot(df.index, df['Signal Line'], label='Signal Line', color='green')
plt.axhline(0, color='grey', linestyle='--', linewidth=1)
plt.title('MACD')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(loc='upper left')
plt.grid(True)

# MACD Histogram 그래프
plt.subplot(3, 1, 3)
plt.bar(df.index, df['MACD'] - df['Signal Line'], label='MACD Histogram', color='grey')
plt.axhline(0, color='grey', linestyle='--', linewidth=1)
plt.title('MACD Histogram')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## RSI (Relative Strength Index) 

```python
# 14일 기간 동안의 가격 변동을 기반으로 RSI를 계산
```

```python
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240601"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터프레임의 인덱스를 날짜 형식으로 변환
df.index = pd.to_datetime(df.index)

# RSI 계산 함수
def calculate_rsi(df, window=14):
    delta = df['종가'].diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    
    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    
    return df

# RSI 계산
df = calculate_rsi(df)

# RSI 시각화
plt.figure(figsize=(12, 8))

# 주가 그래프
plt.subplot(2, 1, 1)
plt.plot(df.index, df['종가'], label='Close Price', color='blue')
plt.title('Samsung Electronics (005930.KS) Close Price and RSI')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)

# RSI 그래프
plt.subplot(2, 1, 2)
plt.plot(df.index, df['RSI'], label='RSI', color='red')
plt.axhline(70, color='grey', linestyle='--')
plt.axhline(30, color='grey', linestyle='--')
plt.title('Relative Strength Index (RSI)')
plt.xlabel('Date')
plt.ylabel('RSI Value')
plt.legend(loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## OBV (On-Balance Volume) 

```python
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240101"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터프레임의 인덱스를 날짜 형식으로 변환
df.index = pd.to_datetime(df.index)

# OBV 계산 함수
def calculate_obv(df):
    df['OBV'] = 0
    for i in range(1, len(df)):
        if df['종가'].iloc[i] > df['종가'].iloc[i - 1]:
            df['OBV'].iloc[i] = df['OBV'].iloc[i - 1] + df['거래량'].iloc[i]
        elif df['종가'].iloc[i] < df['종가'].iloc[i - 1]:
            df['OBV'].iloc[i] = df['OBV'].iloc[i - 1] - df['거래량'].iloc[i]
        else:
            df['OBV'].iloc[i] = df['OBV'].iloc[i - 1]
    return df

# OBV 계산
df = calculate_obv(df)

# OBV 시각화
plt.figure(figsize=(12, 8))

# 주가 그래프
plt.subplot(2, 1, 1)
plt.plot(df.index, df['종가'], label='Close Price', color='blue')
plt.title('Samsung Electronics (005930.KS) Close Price and OBV')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)

# OBV 그래프
plt.subplot(2, 1, 2)
plt.plot(df.index, df['OBV'], label='OBV', color='purple')
plt.title('On-Balance Volume (OBV)')
plt.xlabel('Date')
plt.ylabel('OBV Value')
plt.legend(loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## VR (Volume Ratio)

```python
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240101"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터프레임의 인덱스를 날짜 형식으로 변환
df.index = pd.to_datetime(df.index)

# VR 계산 함수
def calculate_vr(df, period=24):
    df['AV'] = df['거래량'].where(df['종가'] > df['종가'].shift(1), 0)
    df['BV'] = df['거래량'].where(df['종가'] < df['종가'].shift(1), 0)
    df['CV'] = df['거래량'].where(df['종가'] == df['종가'].shift(1), 0)
    
    df['AVS'] = df['AV'].rolling(window=period).sum()
    df['BVS'] = df['BV'].rolling(window=period).sum()
    df['CVS'] = df['CV'].rolling(window=period).sum()
    
    df['VR'] = (df['AVS'] + df['CVS'] / 2) / (df['BVS'] + df['CVS'] / 2) * 100
    return df

# VR 계산
df = calculate_vr(df)

# VR 시각화
plt.figure(figsize=(12, 8))

# 주가 그래프
plt.subplot(2, 1, 1)
plt.plot(df.index, df['종가'], label='Close Price', color='blue')
plt.title('Samsung Electronics (005930.KS) Close Price and Volume Ratio (VR)')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)

# VR 그래프
plt.subplot(2, 1, 2)
plt.plot(df.index, df['VR'], label='VR', color='purple')
plt.axhline(150, color='grey', linestyle='--')
plt.axhline(70, color='grey', linestyle='--')
plt.title('Volume Ratio (VR)')
plt.xlabel('Date')
plt.ylabel('VR Value')
plt.legend(loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## 엘리어트 파동 (사례로)

```python
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt

# 삼성전자 주가 데이터 불러오기
start_date = "20230101"
end_date = "20240101"
ticker = "005930"

df = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터프레임의 인덱스를 날짜 형식으로 변환
df.index = pd.to_datetime(df.index)

# 종가 데이터
close_prices = df['종가']

# 엘리어트 파동의 단순한 예제 좌표 (임의로 설정)
elliott_wave_points = [0, 5, 2, 7, 4, 9, 5, 10]

# 엘리어트 파동의 실제 좌표를 종가 데이터에 맞추기 위해 인덱스를 설정
wave_indices = [0, len(close_prices)//7, len(close_prices)*2//7, len(close_prices)*3//7, 
                len(close_prices)*4//7, len(close_prices)*5//7, len(close_prices)*6//7, len(close_prices)-1]

# 엘리어트 파동 좌표를 실제 종가 데이터에 맞추기
elliott_wave = [close_prices.iloc[i] for i in wave_indices]

# 주가 및 엘리어트 파동 시각화
plt.figure(figsize=(14, 8))
plt.plot(df.index, close_prices, label='Close Price', color='blue')
plt.plot([df.index[i] for i in wave_indices], elliott_wave, marker='o', linestyle='-', color='red', label='Elliott Wave')
plt.title('Samsung Electronics (005930.KS) Close Price with Elliott Wave')
plt.xlabel('Date')
plt.ylabel('Price (KRW)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
